# 2. Research Topic: How do acceptance rates influence enrollment rates in private versus public universities, and what differences can be observed between these two types of institutions?

```python
# Filter relevant columns and create new columns
df = college[['Private', 'Apps', 'Accept', 'Enroll']].copy()
df['Acceptance_Rate'] = df['Accept'] / df['Apps'] * 100
df['Enrollment_Rate'] = df['Enroll'] / df['Accept'] * 100
```

```python
 #Boxplot to compare acceptance rates
plt.figure(figsize=(10, 6))
sns.boxplot(x='Private', y='Acceptance_Rate', data=df)
plt.title('Acceptance Rate Comparison: Private vs. Public Universities')
plt.show()

# Boxplot to compare enrollment rates
plt.figure(figsize=(10, 6))
sns.boxplot(x='Private', y='Enrollment_Rate', data=df)
plt.title('Enrollment Rate Comparison: Private vs. Public Universities')
plt.show()

# Scatter plot to visualize relationship between acceptance and enrollment rates
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Acceptance_Rate', y='Enrollment_Rate', hue='Private', data=df)
plt.title('Acceptance Rate vs. Enrollment Rate')
plt.show()
```

```python
import statsmodels.api as sm

# Separate data for private and public universities
df_private = df[df['Private'] == 'Yes']
df_public = df[df['Private'] == 'No']

# Define the regression function
def regression_analysis(data, title):
    X = data['Acceptance_Rate']
    y = data['Enrollment_Rate']
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    print(f"\n{title}")
    print(model.summary())
    return model

# Perform regression analysis for private universities
model_private = regression_analysis(df_private, 'Private Universities')

# Perform regression analysis for public universities
model_public = regression_analysis(df_public, 'Public Universities')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for the plot
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 8))

# Plot for private universities
sns.regplot(x='Acceptance_Rate', y='Enrollment_Rate', data=df_private,
            scatter_kws={'s': 50, 'alpha': 0.5}, line_kws={'color': 'red'}, label='Private Universities')

# Plot for public universities
sns.regplot(x='Acceptance_Rate', y='Enrollment_Rate', data=df_public,
            scatter_kws={'s': 50, 'alpha': 0.5}, line_kws={'color': 'blue'}, label='Public Universities')

# Add titles and labels
plt.title('Acceptance Rate vs. Enrollment Rate: Private vs. Public Universities')
plt.xlabel('Acceptance Rate (%)')
plt.ylabel('Enrollment Rate (%)')
plt.legend()
plt.show()
```
